```python
import matplotlib.pyplot as plt
from numpy import cov, argsort, maximum, cumsum
from numpy.linalg import eigh
from util import get_MNIST
```

```python
X_train, Y_train, X_test, Y_test = get_MNIST()
```

```python
# Getting covariance and decomposing
cov_X = cov(X_train.T)
lambdas, Q = eigh(cov_X)
```

```python
lambdas
```

```python
# Eigenvalues are in increasing order. Changing to decreasing order
idx = argsort(-lambdas)
lambdas = lambdas[idx]
# Converting negative lambdas to 0 as this is numeric approximation
lambdas = maximum(lambdas, 0)
Q = Q[:, idx]
```

```python
Z = X_train.dot(Q)
plt.scatter(Z[:, 0], Z[:, 1], s = 100, c = Y_train, alpha = 0.3)
plt.show()
# Same plot, just mirrored
```

```python
plt.plot(lambdas)
plt.title("Variance of each component")
plt.show()
```

```python
plt.plot(cumsum(lambdas))
plt.title("Cumulative variance")
plt.show()
```
