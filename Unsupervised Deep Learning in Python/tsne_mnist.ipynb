```python
from util import get_MNIST
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
X, Y, _, _ = get_MNIST()
```

```python
sample_size = 1000 # TSNE crashes otherwise
X = X[:sample_size]
Y = Y[:sample_size]
```

```python
tsne = TSNE()
Z = tsne.fit_transform(X)
plt.scatter(Z[:, 0], Z[:, 1], s = 100, c = Y, alpha = 0.5)
plt.show()
```
