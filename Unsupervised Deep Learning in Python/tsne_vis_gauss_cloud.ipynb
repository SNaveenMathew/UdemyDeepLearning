```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.manifold import TSNE
from util import get_gauss_cloud
from numpy import array
```

```python
pts_per_cloud = 50
centers, dat = get_gauss_cloud()
```

```python
cols = array([[i] * pts_per_cloud for i in range(len(centers))]).flatten()
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(dat[:, 0], dat[:, 1], dat[:, 2], c = cols)
plt.show()
```

```python
tsne = TSNE()
transformed = tsne.fit_transform(dat)
```

```python
plt.scatter(transformed[:, 0], transformed[:, 1], c = cols)
plt.show()
```
